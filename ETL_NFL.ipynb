```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
basic_stats = "basic_stats.csv"
basic_stats_df = pd.read_csv('basic_stats.csv', engine ='python')
basic_stats_df.head()
```

```python
career_stats_passing = "career_stats_passing.csv"
career_stats_passing_df = pd.read_csv('career_stats_passing.csv', engine ='python')
career_stats_passing_df.head()
```

```python
# Create a filtered dataframe from specific columns
basic_stats_cols = ["Player Id", "Name", "Years Played", "Current Team", "Current Status"]
basic_stats_transformed = basic_stats_df[basic_stats_cols].copy()

# Rename the column headers
basic_stats_transformed = basic_stats_transformed.rename(columns={"Player Id": "id",
                                                          "Name": "player_name",
                                                          "Years Played": "years_played",
                                                             "Current Team": "current_team",
                                                             "Current Status": "status"})

# Clean the data by dropping duplicates and setting the index
basic_stats_transformed.drop_duplicates("id", inplace=True)
basic_stats_transformed.set_index("id", inplace=True)

basic_stats_transformed.head()
```

```python
# Create a filtered dataframe from specific columns
career_stats_passing_cols = ["Player Id", "Games Played", "Passes Attempted", "Passes Completed", "Completion Percentage", "Passer Rating"]
career_stats_passing_transformed = career_stats_passing_df[career_stats_passing_cols].copy()

# Rename the column headers
career_stats_passing_transformed = career_stats_passing_transformed.rename(columns={"Player Id": "id",
                                                          "Games Played": "games_played",
                                                          "Passes Attempted": "pass_attempts",
                                                             "Passes Completed": "pass_complete",
                                                             "Completion Percentage": "completion_percentage",
                                                                                    "Passer Rating": "passer_rating"})

# Clean the data by dropping duplicates and setting the index
career_stats_passing_transformed.drop_duplicates("id", inplace=True)
career_stats_passing_transformed.set_index("id", inplace=True)

career_stats_passing_transformed.head()
```

```python
connection_string = "postgres:Jonahdae#13@localhost:5432/nfl_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
basic_stats_transformed.to_sql(name='basic_stats', con=engine, if_exists='append', index=True)
```
